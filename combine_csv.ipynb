```python
# Import 2015 csv files
q4_2015 = pd.read_csv("data/trips/Divvy_Trips_2015_Q4.csv")
m9_2015 = pd.read_csv("data/trips/Divvy_Trips_2015_09.csv")
m8_2015 = pd.read_csv("data/trips/Divvy_Trips_2015_08.csv")
m7_2015 = pd.read_csv("data/trips/Divvy_Trips_2015_07.csv")
q2_2015 = pd.read_csv("data/trips/Divvy_Trips_2015-Q2.csv")
q1_2015 = pd.read_csv("data/trips/Divvy_Trips_2015-Q1.csv")
```

```python
# Combine 2015 files into one df
trips_2015 = pd.concat([q4_2015, m9_2015, m8_2015, m7_2015, q2_2015, q1_2015], ignore_index = True) 
```

```python
# Import 2014 csv files
q4_2014 = pd.read_csv("data/trips/Divvy_Trips_2014-Q4.csv")
q3_0809_2014 = pd.read_csv("data/trips/Divvy_Trips_2014-Q3-0809.csv")
q3_07_2014 = pd.read_csv("data/trips/Divvy_Trips_2014-Q3-07.csv")
q1q2_2014 = pd.read_csv("data/trips/Divvy_Trips_2014_Q1Q2.csv")
```

```python
# Combine 2014 files into one df
trips_2014 = pd.concat([q4_2014, q3_0809_2014, q3_07_2014, q1q2_2014], ignore_index = True) 
```

```python
trips_2014.head()
```

```python
# Import 2013 csv file
trips_2013 = pd.read_csv("data/trips/Divvy_Trips_2013.csv")
```

```python
trips_2013.head()
```

```python
# Find differences in column names
print(trips_2013.columns)
print(trips_2014.columns)
print(trips_2015.columns)
print(trips_2016.columns)
print(trips_2017.columns)
print(trips_2018.columns)
```

```python
# Rename 2013 columns
trips_2013 = trips_2013.rename(
    columns={'starttime':'start_time',
             'stoptime':'end_time',
             'birthday':'birthyear'
              })

# Rename 2014, 2015, and 2016 columns
trips_2014 = trips_2014.rename(
    columns={'starttime':'start_time',
             'stoptime':'end_time'
              })
trips_2015 = trips_2015.rename(
    columns={'starttime':'start_time',
             'stoptime':'end_time'
              })
trips_2016 = trips_2016.rename(
    columns={'starttime':'start_time',
             'stoptime':'end_time'
              })
```

```python
# Combine all years into one df
trips = pd.concat([trips_2013, trips_2014, trips_2015, trips_2016, trips_2017, trips_2018], ignore_index = True)
```

```python
trips["start_time"] = pd.to_datetime(trips["start_time"])
trips["end_time"] = pd.to_datetime(trips["end_time"])
```

```python
trips.dtypes
```

```python
trips.to_csv("data/trips/divvy_trips.csv", index=False, header=True)
```
